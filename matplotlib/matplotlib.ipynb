```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [2,3,4,6,7]
y = [44,55,77,20,80]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10,5))

# 3. Plot Data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Sample Plot",
      xlabel = "x-axis",
      ylabel="y-axis")

# 5. Save fig and show
fig.savefig('sample.png')
```

## Making figures with numpy arrays

we want
* Line plot
* Scatter plot
* Bar plot
* Histogram 
* Subplots


```python
import pandas as pd
import numpy as np

sam = np.linspace(1,10, 100) 

sam
```

```python
fig, ax = plt.subplots()
ax.plot(sam, sam**2)
```

```python
#scatter plot
fig, ax = plt.subplots()
ax.scatter(sam, np.exp(sam))
```

```python
# barplot
fig, ax = plt.subplots()
a = {'almond': 5, 'cashew': 10, 'Peanut':3}
ax.bar(a.keys(), a.values())
ax.set(title="butter prices",
      ylabel="Prices")
```

```python
#Histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Subplots

```python
fig,ax = plt.subplots(nrows=2,
                     ncols=2, figsize = (10,5))
ax[0,0].plot(sam, sam/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(a.keys(),a.values())
ax[1,1].hist(np.random.randn(100))
```

```python
#method 2

fig,((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                     ncols=2, figsize = (10,5))
ax1.plot(sam, sam/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(a.keys(),a.values());
ax4.hist(np.random.randn(100));
```

```python
heart_disease = pd.read_csv('../data/heart-disease.csv')
heart_disease
```

```python
heart_disease.plot.hist(figsize = (10,10), subplots = True);
```

```python
over_50 = heart_disease[heart_disease['age']>50]
over_50
```

```python
# Pyplot method
over_50.plot(kind = 'scatter',x= 'age', y = 'chol', c = 'target');
```

```python
# Matplotlib OO method
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(x = 'age', y = 'chol', kind= 'scatter', c = 'target', ax=ax);
ax.set_xlim([45,100])
```

```python
# Matplotlib OO method
fig, ax = plt.subplots(figsize = (10,6))
scatter = ax.scatter(x = over_50['age'], y = over_50['chol'], c = over_50['target'])

ax.set(title="Heart disease and cholestrol level",
      xlabel='age',
      ylabel='chol')

ax.legend(*scatter.legend_elements(),title='Target')

ax.axhline(over_50['chol'].mean(), linestyle='--')
```

```python
over_50.head()
```

```python
# Matplotlib OO method
fig, (ax0,ax1) = plt.subplots(nrows=2,ncols=1, figsize = (10,6), sharex=True )


#Add data to ax0
scatter = ax0.scatter(x = over_50['age'], y = over_50['chol'], c = over_50['target'])

# Customize ax0
ax0.set(title="Heart disease and cholestrol level",
#       xlabel='age',
      ylabel='chol')

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title='Target')

# Add a mean line
ax0.axhline(over_50['chol'].mean(), linestyle='--')



scatter = ax1.scatter(x = over_50['age'], y = over_50['thalach'], c = over_50['target'])

ax1.set(title="Heart disease and cholestrol level",
      xlabel='age',
      ylabel='thalach')

ax1.legend(*scatter.legend_elements(),title='Target')

ax1.axhline(over_50['thalach'].mean(), linestyle='--')

fig.suptitle("Heart Disease Analysis", fontsize = 15, fontweight = 'bold')
```
